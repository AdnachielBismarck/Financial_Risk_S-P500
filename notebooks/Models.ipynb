# Mathematical and Computational Foundations of the Models

---

## 1. Introduction

This project aims to model the occurrence of extreme financial events through a combination of stochastic processes and volatility modeling. Instead of predicting prices directly, the approach focuses on **risk representation**, capturing both the magnitude and frequency of extreme market movements.

---

## 2. Motivation for Model Selection

Financial time series exhibit well-known stylized facts:

- Volatility clustering  
- Heavy tails  
- Time-varying variance  

Traditional linear models fail to capture these characteristics. Therefore, models are selected based on their ability to represent these empirical properties.

---

## 3. GARCH(1,1) Model

### 3.1 Mathematical Formulation

The GARCH(1,1) model defines conditional variance as:

$$
\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2
$$

where:

- $\omega > 0$ is a constant
- $\alpha$ captures the impact of recent shocks
- $\beta$ captures volatility persistence

---

### 3.2 Interpretation

- High $\alpha$: strong reaction to market shocks  
- High $\beta$: long memory in volatility  

A key condition:

$$
\alpha + \beta \approx 1
$$

indicates **volatility clustering**, a well-documented empirical feature of financial markets.

---

### 3.3 Justification

GARCH is selected because:

- It models **heteroskedasticity**
- Captures volatility clustering
- Provides a dynamic estimate of risk

---

### 3.4 Computational Perspective

The model is estimated via Maximum Likelihood Estimation (MLE), assuming a conditional normal distribution of returns.

Scaling returns (multiplying by 100) improves numerical stability during optimization.

---

## 4. Extreme Event Definition

Extreme events are defined as:

$$
Y_t =
\begin{cases}
1 & \text{if } r_t < q_{0.05} \\
0 & \text{otherwise}
\end{cases}
$$

where $q_{0.05}$ is the 5th percentile of returns.

This approach ensures a data-driven threshold rather than an arbitrary cutoff.

---

## 5. Non-Homogeneous Poisson Process

### 5.1 Model Definition

The occurrence of extreme events is modeled as a Poisson process with time-varying intensity:

$$
\lambda(t) = \exp(\beta_0 + \beta_1 \sigma_t)
$$

where:

- $\sigma_t$ is the volatility (from GARCH)
- $\lambda(t)$ represents the expected event rate

---

### 5.2 Interpretation

- Higher volatility → higher event intensity  
- $\lambda(t)$ acts as a **risk indicator**

---

### 5.3 Justification

The Poisson model is appropriate because:

- Events are discrete (0 or 1 per day)
- Focus is on **event frequency**, not magnitude
- It allows interpretability via intensity function

---

### 5.4 Computational Perspective

The model is implemented using a Generalized Linear Model (GLM) with:

- Poisson family  
- Log link function  

This ensures:

$$
\lambda(t) > 0
$$

---

## 6. Integration of Models

The pipeline is:

1. Compute returns  
2. Estimate volatility via GARCH  
3. Use volatility as input for Poisson model  

This creates a dependency structure:

$$
\text{Volatility} \rightarrow \text{Event Intensity}
$$

---

## 7. Limitations

- GARCH assumes symmetry in volatility response  
- Poisson assumes conditional independence  
- Extreme event definition depends on threshold selection  

---

## 8. Conclusion

The combination of GARCH and a non-homogeneous Poisson process provides:

- A dynamic measure of risk  
- Interpretability  
- Alignment with financial theory  

This approach moves beyond price prediction and focuses on **risk modeling**, which is more relevant for decision-making in financial contexts.
